```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Análise formulário pré-aula

```python
pre = pd.read_excel("pre.xlsx")
```

```python
pre.head()
```

```python
pre = pre.drop("ID", axis=1)
pre = pre.drop("Hora de início", axis=1)
pre = pre.drop("Hora de conclusão", axis=1)
pre = pre.drop("Email", axis=1)
pre = pre.drop("Nome", axis=1)
pre = pre.drop("Hora da última modificação", axis=1)
```

```python
controle1 = pre[pre["Qual a sua turma?"] == "Controle 1"]
controle2 = pre[pre["Qual a sua turma?"] == "Controle 2"]
experimental = pre[pre["Qual a sua turma?"] == "Experimental"]
```

```python
experimental.head()
```

```python
likert_map = {
    "Discordo totalmente": 1,
    "Discordo parcialmente": 2,
    "Neutro": 3,
    "Concordo parcialmente": 4,
    "Concordo totalmente": 5,
    "Nenhum": 1,
    "Básico": 2,
    "Intermediário": 3,
    "Avançado": 4,
    "Especialista": 5
}
```

```python
colsExperimental = experimental.columns[2:10]

experimental = experimental.copy()

for c in colsExperimental:
    experimental[c] = experimental[c].map(likert_map)

mediaExperimental = experimental[colsExperimental].mean().to_dict()
medianaExperimental = experimental[colsExperimental].median().to_dict()
modaExperimental = experimental[colsExperimental].mode().iloc[0].to_dict()
desvioExperimental = experimental[colsExperimental].std().to_dict()

summaryExperimental = pd.DataFrame({
    "Média": experimental[colsExperimental].mean().round(2),
    "Mediana": experimental[colsExperimental].median().round(2),
    "Moda": experimental[colsExperimental].mode().iloc[0].round(2),
    "Desvio Padrão": experimental[colsExperimental].std().round(2)
})

print(summaryExperimental)
```

```python
porcentagensExperimental = experimental[colsExperimental].apply(lambda x: (x.value_counts(normalize=True) * 100).round(1))
print(porcentagensExperimental)
```

```python
colsControle1 = controle1.columns[2:10]

controle1 = controle1.copy()

for c in colsControle1:
  controle1[c] = controle1[c].map(likert_map)

mediaControle1 = controle1[colsControle1].mean().to_dict()
medianaControle1 = controle1[colsControle1].median().to_dict()
modaControle1 = controle1[colsControle1].mode().iloc[0].to_dict()
desvioControle1 = controle1[colsControle1].std().to_dict()

summaryControle1 = pd.DataFrame({
    "Média": controle1[colsControle1].mean().round(2),
    "Mediana": controle1[colsControle1].median().round(2),
    "Moda": controle1[colsControle1].mode().iloc[0].round(2),
    "Desvio Padrão": controle1[colsControle1].std().round(2)
})

print(summaryControle1)
```

```python
porcentagensControle1 = controle1[colsControle1].apply(lambda x: (x.value_counts(normalize=True) * 100).round(1))
print(porcentagensControle1)
```

```python
colsControle2 = controle2.columns[2:10]

controle2 = controle2.copy()

for c in colsControle2:
  controle2[c] = controle2[c].map(likert_map)

mediaControle2 = controle2[colsControle2].mean().to_dict()
medianaControle2 = controle2[colsControle2].median().to_dict()
modaControle2 = controle2[colsControle2].mode().iloc[0].to_dict()
desvioControle2 = controle2[colsControle2].std().to_dict()

summaryControle2 = pd.DataFrame({
    "Média": controle2[colsControle2].mean().round(2),
    "Mediana": controle2[colsControle2].median().round(2),
    "Moda": controle2[colsControle2].mode().iloc[0].round(2),
    "Desvio Padrão": controle2[colsControle2].std().round(2)
})

print(summaryControle2)
```

```python
porcentagensControle2 = controle2[colsControle2].apply(lambda x: (x.value_counts(normalize=True) * 100).round(1))
print(porcentagensControle2)
```

```python
comparativo = pd.DataFrame({
    "Controle 1": summaryControle1["Média"],
    "Controle 2": summaryControle2["Média"],
    "Experimental": summaryExperimental["Média"]
})
```

```python
comparativo.plot(kind="bar", figsize=(10,6))
plt.title("Comparativo de Médias por Pergunta", fontsize=16)
plt.ylabel("Média (1 a 5)", fontsize=16)
plt.xlabel("Perguntas", fontsize=16)
plt.xticks(range(8), ["Q3","Q4","Q5","Q6","Q7","Q8","Q9","Q10"],
           rotation=0, ha="right", fontsize=16)
plt.yticks(fontsize=16)
plt.legend(title="Turma", fontsize=12, title_fontsize=12)
plt.tight_layout()
plt.show()
```

```python
desvios = pd.DataFrame({
    "Controle 1": summaryControle1["Desvio Padrão"],
    "Controle 2": summaryControle2["Desvio Padrão"],
    "Experimental": summaryExperimental["Desvio Padrão"]
})
```

```python
# Perguntas (eixo X)
perguntas = summaryControle1.index

# Valores médios
valores_controle1 = summaryControle1["Média"]
valores_controle2 = summaryControle2["Média"]
valores_experimental = summaryExperimental["Média"]

# Desvios padrão
dp_controle1 = summaryControle1["Desvio Padrão"]
dp_controle2 = summaryControle2["Desvio Padrão"]
dp_experimental = summaryExperimental["Desvio Padrão"]

# Cria deslocamento no eixo X para visualizar as turmas lado a lado
x = np.arange(len(perguntas))
largura = 0.1

plt.figure(figsize=(12,9))

# Gráfico de dispersão com barras de erro
plt.errorbar(x - largura, valores_controle1, yerr=dp_controle1, fmt='o', capsize=5, label="Controle 1", linewidth=4)
plt.errorbar(x, valores_controle2, yerr=dp_controle2, fmt='s', capsize=5, label="Controle 2", linewidth=4)
plt.errorbar(x + largura, valores_experimental, yerr=dp_experimental, fmt='^', capsize=5, label="Experimental", linewidth=4)

plt.ylabel("Pontuação (1 a 5)", fontsize=20)
plt.xlabel("Perguntas", fontsize=20)
plt.title("Gráfico de dispersão com desvio padrão por turma", fontsize=20)
plt.yticks(fontsize=20)
plt.xticks(range(6), ["Q3","Q4","Q5","Q6","Q7","Q8"],
           rotation=0, ha="right", fontsize=20)
plt.legend(title="Turma", fontsize=20, title_fontsize=20)
plt.tight_layout()
plt.show()
```

```python
df_todas.head()
```

```python
# Contar alunos por curso em cada turma
contagem = df_todas.groupby(['Turma', 'Qual é o seu curso de graduação?']).size().unstack(fill_value=0)

# Gráfico de barras empilhadas
contagem.plot(kind='bar', stacked=True, figsize=(10,5), colormap='Set3')
plt.title("Número de alunos por curso em cada turma", fontsize=14)
plt.ylabel("Número de alunos", fontsize=14)
plt.xlabel("Turma", fontsize=14)
plt.xticks(rotation=0, fontsize=14)
plt.yticks(fontsize=14)
plt.legend(title="Curso", title_fontsize = 14, bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=14)
plt.tight_layout()
plt.show()
```
